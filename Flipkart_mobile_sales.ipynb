```python
df=df.drop(columns=['Original Price'])
```

```python
df_clean=df.dropna(axis=0)
#axis=0 is for row and axis=1 is for columns
#By this above code, we are dropping out rows which contain NaN values
```

```python
df_clean.shape
```

```python
df_clean.head()
```

```python
brands=df_clean.groupby(by='Brand').count().sort_values(ascending=False,by='Model')
```

```python
brands.head()
```

```python
brands=brands.reset_index()
```

```python
brands.head()
```

```python
fig=px.bar(brands,x='Model',y='Brand',color='Brand', title='Number of Models by brand')
fig.show()
```

```python
df=df_clean
```

Now, in color column we'll store the general color. General means simply black,green instead of what we already have - electric black, mint green ,etc

```python
df['Color']=df['Color'].str.split(n=1).str[-1] #Yaha reverse indexing se last splitted string ko access kiye h.
df.head()
```

```python
Colors=df.groupby('Color').count().sort_values(ascending=False,by='Model')
Colors=Colors.reset_index()
Colors.head()
```

```python
color_10=Colors.sort_values(ascending=False,by=['Model'])[:10] #top 10 Colors according to model
color_10.reset_index()
```

```python
fig=px.bar(color_10,x='Color',y='Model',color='Color',title='Top 10 in-demand colors')
fig.show()
```

```python
ratings=df.groupby('Brand')[['Rating']].mean().sort_values(ascending=False,by='Rating').reset_index()
ratings.head()
```

```python
fig=px.bar(ratings,x='Rating',y='Brand',color='Brand',title='Avg rating by Brands')
fig.show()
```

```python
fig=px.box(df,x='Brand',y='Selling Price', title='Selling Price per Brand', color='Brand')
fig.show()
```

```python
df['Segments']=pd.cut(df['Selling Price'],[0,25000,50000,np.inf],labels=['Budget','Midrange','Flagship'])
#see documentation OR video from duration 1:00:00
df.head()
```

```python
segments=df.groupby(['Segments']).count().reset_index()
segments
```

```python
fig=px.pie(segments,values='Model',names='Segments',title='Phones per segment')
fig.show()
```

```python
fig=px.scatter(df,x='Rating',y='Selling Price', color='Brand')
fig.show()
```

```python
fig=px.scatter(df,x='Rating',y='Segments',color='Segments')
fig.show()
```
